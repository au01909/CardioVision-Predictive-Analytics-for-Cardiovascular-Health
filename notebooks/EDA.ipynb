# CardioVision - Exploratory Data Analysis

This notebook explores the UCI Heart Disease dataset used to train CardioVision's risk prediction model.

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import config
import preprocessing

sns.set_style('whitegrid')
%matplotlib inline
```

## 1. Load & Inspect the Dataset

```python
df = preprocessing.load_dataset()
df.head()
```

```python
df.describe()
```

## 2. Missing Values

```python
df.isna().sum()
```

## 3. Target Distribution

```python
sns.countplot(x=config.TARGET_COL, data=df)
plt.title('Target Distribution')
plt.show()
df[config.TARGET_COL].value_counts(normalize=True)
```

## 4. Correlation Matrix

```python
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Feature Correlation Matrix')
plt.show()
```

## 5. Feature Distributions & Outliers

```python
fig, axes = plt.subplots(2, 3, figsize=(16, 9))
for ax, col in zip(axes.flatten(), config.NUMERIC_FEATURES):
    sns.histplot(df[col], kde=True, ax=ax)
    ax.set_title(col)
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(16, 9))
for ax, col in zip(axes.flatten(), config.NUMERIC_FEATURES):
    sns.boxplot(y=df[col], ax=ax)
    ax.set_title(col)
plt.tight_layout()
plt.show()
```
